##  playing a video backwards 

```python
import cv2
import numpy as np

capture=cv2.VideoCapture('cars.avi')

if capture.isOpened() is False:
    print("error while opening the video")
    
frame_index=capture.get(cv2.CAP_PROP_FRAME_COUNT)-1

print(f'starting frame is {frame_index}')

while capture.isOpened() and frame_index>=0:
    
    capture.set(cv2.CAP_PROP_POS_FRAMES,frame_index)
    
    ret,frame=capture.read()
    
    if ret is True:
        
        cv2.imshow("inputframe",frame)
        grayframe=cv2.cvtColor(frame,cv2.COLOR_BGR2GRAY)
        cv2.imshow("gray frame",grayframe)
        frame_index -=1
        print(f"upcoming frame is {frame_index}")
        
    else:
        break
        
    if cv2.waitKey(10) & 0xFF==ord('m'):
        break
        
capture.release()
cv2.destroyAllWindows()
```
