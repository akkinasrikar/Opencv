```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
colors = {'red': (255, 0, 0),
          'green': (0, 255, 0),
          'blue': (0, 0, 255),
          'yellow': (0, 255, 255),
          'magenta': (255, 0, 255),
          'light_gray': (220, 220, 220)}
image=np.zeros((300,300,3),np.uint8)
plt.figure(figsize=(10,6))
plt.imshow(image)
plt.show()
```

```python
cv2.line(image,(0,0),(300,300),colors['green'],4)
cv2.rectangle(image,(0,0),(150,150),colors['blue'],4)
revt,p1,p2=cv2.clipLine((0,0,150,150),(0,0),(300,300))
if revt is True:
    cv2.line(image,p1,p2,colors['yellow'],4)
    
plt.figure(figsize=(10,6))
plt.imshow(image)
```

```python
image=np.zeros((300,300,3),np.uint8)
```

```python
cv2.arrowedLine(image,(50,50),(250,250),colors['red'],3,tipLength=0.3)
plt.figure(figsize=(10,6))
plt.imshow(image)
```

```python
image=np.zeros((300,300,3),np.uint8)
```

```python
cv2.ellipse(image,(80,80),(60,40),0,0,360,colors['green'],-1)
plt.figure(figsize=(10,6))
plt.imshow(image)
```

```python
image=np.zeros((300,300,3),np.uint8)
cv2.ellipse(image, (80, 80), (60, 40), 0, 0, 360, colors['red'], -1)
cv2.ellipse(image, (80, 200), (80, 40), 0, 0, 360, colors['green'], 3)
cv2.ellipse(image, (80, 200), (10, 40), 0, 0, 360, colors['blue'], 3)
cv2.ellipse(image, (200, 200), (10, 40), 0, 0, 180, colors['yellow'], 3)
cv2.ellipse(image, (250, 250), (30, 30), 0, 0, 360, colors['magenta'], 3)
plt.figure(figsize=(10,6))
plt.imshow(image)
```

```python
image=np.zeros((500,500,3),np.uint8)
pts=np.array([[10,50],[400,50],[90,200],[50,500]],np.int32)
pts.reshape(4,1,2)
cv2.polylines(image,[pts],True,colors['blue'],4)
plt.imshow(image)
```
