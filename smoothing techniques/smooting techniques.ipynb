```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
def display_image(image,title):
    image=image[:,:,::-1]
    plt.figure(figsize=(10,6))
    plt.imshow(image)
    plt.title(title)
    plt.show()
```

```python
image=cv2.imread("ronoldo.jpg")
display_image(image,"original image")
```

```python
kernel_averaging_10_10=np.ones((10,10),np.float32)/100
kernel_averaging_5_5=np.ones((5,5),np.float32)/100
```

```python
smooth_image_5_5=cv2.filter2D(image,-1,kernel_averaging_5_5)
smooth_image_10_10=cv2.filter2D(image,-1,kernel_averaging_10_10)
smooth_image_blur=cv2.blur(image,(10,10))
smooth_image_boxfilter=cv2.boxFilter(image,-1,(10,10))
smooth_image_gaussianblur=cv2.GaussianBlur(image,(7,7),2)
smooth_image_median_blur=cv2.medianBlur(image,9)
"""
.   _Sigma values_: For simplicity, you can set the 2 sigma values to be the same. If they are small (\<
.   10), the filter will not have much effect, whereas if they are large (\> 150), they will have a very
.   strong effect, making the image look "cartoonish".
.   
.   _Filter size_: Large filters (d \> 5) are very slow, so it is recommended to use d=5 for real-time
.   applications, and perhaps d=9 for offline applications that need heavy noise filtering.
"""
smooth_image_bilateralfilter1=cv2.bilateralFilter(image,5,10,10)
smooth_image_bilateralfilter2=cv2.bilateralFilter(image,9,200,200)
```

```python
display_image(smooth_image_5_5,"smooth_image_5_5")
display_image(smooth_image_10_10,"smooth_image_10_10")
display_image(smooth_image_blur,"smooth_image_blur")
display_image(smooth_image_boxfilter,"smooth_image_boxfilter")
display_image(smooth_image_gaussianblur,"smooth_image_gaussianblur")
display_image(smooth_image_median_blur,"smooth_image_median_blur")
display_image(smooth_image_bilateralfilter1,"smooth_image_bilateralfilter1")
display_image(smooth_image_bilateralfilter2,"smooth_image_bilateralfilter2")
```
