```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
def display_image(image,title):
    image=image[:,:,::-1]
    plt.imshow(image)
    plt.title(title)
    plt.show()
```

```python
image=cv2.imread("lena_image.png")
display_image(image,"original image")
```

```python
resizeimg1=cv2.resize(image,None,fx=0.5,fy=0.5,interpolation=cv2.INTER_AREA)
display_image(resizeimg1,"resizeimg1")
```

```python
height,width=image.shape[:2]
```

```python
resizeimg2=cv2.resize(image,(width*2,height*4),interpolation=cv2.INTER_CUBIC)
display_image(resizeimg2,"resizeimg2")
```

```python
M=np.float32([[1,0,200],[0,1,30]])
transaltedimg1=cv2.warpAffine(image,M,(width,height))
display_image(transaltedimg1,"translatedimg1")
```

```python
M=np.float32([[1,0,-200],[0,1,-30]])
transaltedimg2=cv2.warpAffine(image,M,(width,height))
display_image(transaltedimg2,"translatedimg2")
```

```python
M=cv2.getRotationMatrix2D((width/2,height/2),180,1)
rotatedimg=cv2.warpAffine(image,M,(width,height))
display_image(rotatedimg,"180deg")
```

```python
M=cv2.getRotationMatrix2D((width/1.5,height/1.5),50,1)
rotatedimg=cv2.warpAffine(image,M,(width,height))
display_image(rotatedimg,"newrotated")
```

```python
cv2.circle(rotatedimg,(round(width / 1.5), round(height / 1.5)), 5, (255, 0, 0), -1)
display_image(rotatedimg,"circle")
```

```python
#affine transformation
image_points=image.copy()
cv2.circle(image_points, (135, 45), 5, (255, 0, 255), -1)
cv2.circle(image_points, (385, 45), 5, (255, 0, 255), -1)
cv2.circle(image_points, (135, 230), 5, (255, 0, 255), -1)
```

```python
points1=np.float32([[135, 45], [385, 45], [135, 230]])
points2=np.float32([[135, 45], [385, 45], [150, 230]])
```

```python
M=cv2.getAffineTransform(points1,points2)
img=cv2.warpAffine(image_points,M,(width,height))
display_image(image,"original")
display_image(img,"affine")
```

```python
image_points = image.copy()
cv2.circle(image_points, (450, 65), 5, (255, 0, 255), -1)
cv2.circle(image_points, (517, 65), 5, (255, 0, 255), -1)
cv2.circle(image_points, (431, 164), 5, (255, 0, 255), -1)
cv2.circle(image_points, (552, 164), 5, (255, 0, 255), -1)
```

```python
points1 = np.float32([[450, 65], [517, 65], [431, 164], [552, 164]])
points2= np.float32([[0, 0], [300, 0], [0, 300], [300, 300]])
```

```python
M=cv2.getPerspectiveTransform(points1,points2)
img=cv2.warpPerspective(image_points,M,(300,300))
display_image(image,"original")
display_image(img,"paffine")
```

```python
#cropping image
image_points=image.copy()
cv2.circle(image_points, (230, 80), 5, (0, 0, 255), -1)
cv2.circle(image_points, (330, 80), 5, (0, 0, 255), -1)
cv2.circle(image_points, (230, 200), 5, (0, 0, 255), -1)
cv2.circle(image_points, (330, 200), 5, (0, 0, 255), -1)
cv2.line(image_points, (230, 80), (330, 80), (0, 0, 255))
cv2.line(image_points, (230, 200), (330, 200), (0, 0, 255))
cv2.line(image_points, (230, 80), (230, 200), (0, 0, 255))
cv2.line(image_points, (330, 200), (330, 80), (0, 0, 255))
```

```python
crpimg=image[80:200, 230:330]
display_image(image_points,"original")
display_image(crpimg,"cropped")
```
